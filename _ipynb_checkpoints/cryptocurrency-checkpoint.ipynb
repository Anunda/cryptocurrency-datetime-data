```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(mydf.index,mydf['Open'],c='red', label='Open')
ax.set_xlabel('Time')
ax.set_ylabel('Open',c='red')
ax.tick_params('y',colors='red')
ax2 = ax.twinx()
ax2.plot(mydf.index,mydf['Volume'],c='b',label='Volume')
ax2.set_ylabel('Volume',c='b')
ax2.tick_params('y',colors='b')
plt.show()
```

```python
def cropvol(axes,x,y,xlabel,ylabel,color):
    axes.plot(x,y,color=color)
    axes.set_xlabel(xlabel)
    axes.set_xticklabels(x, rotation=45)
    axes.set_ylabel(ylabel,color=color)
    axes.tick_params('y',colors=color)
```

This is an example of how to convert visualization to functions
Annotation of text in graphs is also shown here

```python
fig, ax = plt.subplots(figsize=(12,10))
ax2 = ax.twinx()
cropvol(ax,mydf.index,mydf['Volume'],'Time','Volume','red')
cropvol(ax2,mydf.index,mydf['Low'],'Time','Low','blue')
ax.annotate('almost 60m', xy=(pd.Timestamp('2017-12-13'),59491228.13),xytext=(pd.Timestamp('2017-08-01'),70000000),
            arrowprops={'arrowstyle':'->','color':'green'})
plt.show()
```
